```python
import pandas as pd
import numpy as np

%matplotlib inline
from matplotlib.pylab import rcParams
rcParams["figure.figsize"] = 16 , 4
```

```python
from AJ_ML_Library import models_classification
```

```python
train = pd.read_csv("data/train_modified_1.csv")
test = pd.read_csv("data/test_modified_1.csv")
```

```python
train.shape , test.shape
```

```python
train["fault_severity"].value_counts()
```

```python
target = "fault_severity"
IDcol = "id"
output_colnames = ["predict_0" , "predict_1" , "predict_2"]
```

# XGBoost

```python
models_classification = reload(models_classification)
```

```python
predictors = [x for x in train.columns if x not in [target , IDcol]]
```

## Get the num_boost_round for 0.1 learning rate

```python
xgb1 = models_classification.XGBoost_Class(train , test , target , predictors , cv_folds=5 , 
                                          scoring_metric_skl="log_loss" , scoring_metric_xgb="mlogloss")
```

```python
xgb1.set_parameters({
    "learning_rate":0.1 ,
    "n_estimators":500 , 
    "max_depth":10 , 
    "min_child_weight":30 , 
    "gamma":0 , 
    "subsample":0.8 , 
    "colsample_bytree":0.8 ,
    "objective":"multi:softprob",
    "num_class":3 ,
    "eval_metric":["mlogloss"] , 
    "missing":-1 , 
    "nthread":8
})
```

```python
xgb1.modelfit(useTrainCV=True , performCV=False)
```

```python
xgb2 = models_classification.XGBoost_Class(train , test , target , predictors , cv_folds=5 , 
                                          scoring_metric_skl="log_loss" , scoring_metric_xgb="mlogloss")
```

```python
xgb2.set_parameters({
    "learning_rate":0.1,
    "n_estimators":96 , 
    "max_depth":10 , 
    "min_child_weight":30 , 
    "gamma":0,
    "subsample":0.8 , 
    "colsample_bytree":0.8 , 
    "objective":"multi:softprob",
    "num_class":3,
    "eval_metric":["nlogloss"],
    "missing":-1 , 
    "nthread":8
})
```

```python
param_test1 = {
    "max_depth":range(8 , 17 , 2) , 
    "min_child_weight":range(10 , 51 , 10)
}
```

```python
xgb2.GridSearch(param_test1 , n_jobs=8 , iid=False , cv=5)
```
